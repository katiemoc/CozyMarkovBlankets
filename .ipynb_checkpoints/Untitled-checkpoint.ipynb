```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from hmmlearn.hmm import GaussianHMM
import matplotlib.pyplot as plt
import seaborn as sns
# Import necessary library for state mapping
from scipy.optimize import linear_sum_assignment 


df = pd.read_csv('student_sleep_patterns.csv')

features = [
    'Sleep_Duration', 'Study_Hours', 'Screen_Time',
    'Caffeine_Intake', 'Physical_Activity'
]

def map_quality_to_3_states(quality):
    if quality <= 3:
        return 0 # Low
    elif quality <= 7:
        return 1 # Medium
    else:
        return 2 # High

df['Sleep_Quality_3_States'] = df['Sleep_Quality'].apply(map_quality_to_3_states)

X_raw = df[features].values
L = np.array([len(X_raw)])

scaler = StandardScaler()
X = scaler.fit_transform(X_raw)
y = df['Sleep_Quality_3_States'].values



def find_optimal_mapping(ground_truth, decoded_states, n_states):
    # confusion matrix
    cm = confusion_matrix(ground_truth, decoded_states, labels=range(n_states))
    
    # Maximize matches -> Minimize (Total - Matches)
    # cost is negative of the confusion matrix
    row_ind, col_ind = linear_sum_assignment(cm.max() - cm)
    
    # optimal accuracy is sum of the correctly assigned states
    optimal_accuracy = cm[row_ind, col_ind].sum() / len(ground_truth)
    
    # create the mapping dict: model_state -> ground_truth_state
    mapping = {model_state: ground_truth_state for model_state, ground_truth_state in zip(col_ind, row_ind)}
    
    # use the optimal mapping to align the decoded states
    aligned_states = np.array([mapping[state] for state in decoded_states])
    
    return optimal_accuracy, aligned_states


results = {}


def train_and_evaluate_hmm(X, L, n_states, n_iter, ground_truth):
    model = GaussianHMM(
        n_components=n_states,
        covariance_type="full",
        n_iter=n_iter,
        random_state=42 # for reproducibility
    )
    model.fit(X, L)

    #decode
    log_prob, decoded_states = model.decode(X)
    
    # accuracy calculation
    accuracy, aligned_states = find_optimal_mapping(ground_truth, decoded_states, n_states)
    
    # Mean Absolute Error (MAE) using the ALIGNED states
    mae = np.mean(np.abs(ground_truth - aligned_states)) 

    print(f"  Viterbi Accuracy (Aligned): {accuracy:.3f}, MAE (Aligned): {mae:.3f}")

    return {
        'accuracy': accuracy,
        'mae': mae,
        'transmat': model.transmat_,
        'n_states': n_states,
        'model': model
    }

# --- Train/Store Results for the Ablations ---
# Model 1: Initial Model (using 3 states)
results['2 Iterations'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=2, ground_truth=y
)

results['3 Iterations'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=3, ground_truth=y
)

results['4 Iterations'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=4, ground_truth=y
)

results['5 Iterations'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=5, ground_truth=y
)

results['6 Iterations'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=6, ground_truth=y
)

results['7 Iterations'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=7, ground_truth=y
)

results['Initial Model (3 States)'] = train_and_evaluate_hmm(
    X, L, n_states=3, n_iter=200, ground_truth=y
)

df['Sleep_Quality_2_States'] = df['Sleep_Quality'].apply(lambda q: 0 if q <= 5 else 1)
y_2 = df['Sleep_Quality_2_States'].values

results['Ablation: 2 States'] = train_and_evaluate_hmm(
    X, L, n_states=2, n_iter=200, ground_truth=y_2
)

print("\n--- Model Training Complete. Proceeding to Visualizations. ---")


# --- 3. VISUALIZATION 1: PERFORMANCE COMPARISON ---

# Prepare data for the bar chart
data_list = []
for config, res in results.items():
    data_list.append({
        'Model Configuration': config,
        'Metric': 'Viterbi Accuracy',
        'Value': res['accuracy']
    })
    data_list.append({
        'Model Configuration': config,
        'Metric': 'Mean Absolute Error (MAE)',
        'Value': res['mae']
    })

results_df = pd.DataFrame(data_list)

plt.figure(figsize=(12, 6))
sns.barplot(
    x='Model Configuration',
    y='Value',
    hue='Metric',
    data=results_df,
    palette=['#3498db', '#e74c3c']
)

plt.title('Performance Comparison Across HMM Ablation Experiments', fontsize=16)
plt.ylabel('Score')
plt.xlabel('Model Configuration')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Evaluation Metric')
plt.tight_layout()
plt.show()


# --- 4. VISUALIZATION 2: TPM HEATMAPS (Comparing Initial vs. Higher Iterations) ---

# Note: We compare two 3-state models here. If you want to compare 
# a 3-state model vs. a 2-state model, you would need different subplots 
# due to the differing matrix sizes.

tpm_initial = results['Initial Model (3 States)']['transmat']
tpm_recent = results['Ablation: Higher Iterations (500)']['transmat']
state_labels_3 = ['Low (0)', 'Medium (1)', 'High (2)']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
plt.suptitle('Comparison of Learned HMM Transition Probability Matrices (TPM)', fontsize=16, y=1.02)

# Heatmap for Initial Model
sns.heatmap(
    tpm_initial,
    annot=True,
    fmt=".3f", # Show three decimal places
    cmap="Blues",
    linewidths=.5,
    linecolor='white',
    ax=axes[0],
    cbar=True,
    xticklabels=state_labels_3,
    yticklabels=state_labels_3
)
axes[0].set_title('Initial Model TPM (n=3 States, 200 Iterations)')
axes[0].set_xlabel('To State $S_t$')
axes[0].set_ylabel('From State $S_{t-1}$')

# Heatmap for Ablated Model (Higher Iterations)
sns.heatmap(
    tpm_recent,
    annot=True,
    fmt=".3f",
    cmap="Blues",
    linewidths=.5,
    linecolor='white',
    ax=axes[1],
    cbar=True,
    xticklabels=state_labels_3,
    yticklabels=state_labels_3
)
axes[1].set_title('Ablation TPM (n=3 States, 500 Iterations)')
axes[1].set_xlabel('To State $S_t$')
axes[1].set_ylabel('From State $S_{t-1}$')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
